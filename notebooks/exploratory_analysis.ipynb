Exploratory analysis of structural (DTI) and functional (fMRI) brain connectivity matrices, network metrics, and inter-individual variability.

````python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns, warnings
warnings.filterwarnings('ignore'); sns.set_theme(style='whitegrid'); print('Libraries loaded')
````

````python
import os
df=None
for p in ['../data/connectivity_metrics.csv','../data/data.csv']:
    if os.path.exists(p): df=pd.read_csv(p); break
if df is None:
    np.random.seed(42)
    n=30
    df=pd.DataFrame({'subject_id':[f'SUB_{i:03d}' for i in range(n)],'age':np.random.randint(20,65,n),'sex':np.random.choice(['M','F'],n),'global_efficiency':np.random.normal(0.58,0.05,n),'clustering_coef':np.random.normal(0.42,0.06,n),'path_length':np.random.normal(2.31,0.18,n),'modularity':np.random.normal(0.45,0.07,n),'structural_strength':np.random.normal(0.61,0.08,n)})
    print('Using simulated connectomics data')
print(df.shape); df.head()
````

````python
fig,axes=plt.subplots(2,2,figsize=(12,10))
metrics=['global_efficiency','clustering_coef','path_length','modularity']
for ax,m in zip(axes.flat,metrics):
    if m in df.columns:
        ax.hist(df[m],bins=10,color='steelblue',edgecolor='white',alpha=0.8)
        ax.set_title(m.replace('_',' ').title()); ax.set_xlabel('Value'); ax.set_ylabel('Count')
        ax.axvline(df[m].mean(),color='red',linestyle='--',label=f'Mean: {df[m].mean():.3f}')
        ax.legend(fontsize=8)
plt.suptitle('Brain Network Metrics Distribution', fontsize=14, fontweight='bold')
plt.tight_layout(); plt.show()
````
